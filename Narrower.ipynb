# Narrower
### In this notebook, we convert our wide dataframe of confidence levels for features of various dataframes to a narrow format
Import libraries

```python
import pandas as pd
```

Read in our dataframe of confidence levels

```python
cville = pd.read_csv('cville_conf.csv')
```

Check the first few lines

```python
cville.head()
```

Create dictionary of variable names to our description(s)

```python
keys = cville.columns[1:]
values = cville.iloc[0].tolist()[1:]


var_des = dict(zip(keys, values))
```

### Reformatting

Create narrow dataframe

```python
narrow = pd.melt(cville.iloc[1:,:], id_vars=['Variables']).dropna().sort_values('Variables')
narrow.head()
```

Add column for descriptions (from dictionary)

```python
narrow['Description'] = narrow['variable'].map(var_des) 
narrow.head()
```

Prettify column names

```python
narrow.columns = ['Source', 'Variable', 'Confidence', 'Description']
narrow.head()
```

```python
narrow.to_csv('Narrow.csv', index = False)
```
